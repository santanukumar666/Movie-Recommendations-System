```python
movies = pd.read_csv(r'C:\Users\subra\OneDrive\Desktop\datasets\tmdb_5000_movies.csv')
credits = pd.read_csv(r'C:\Users\subra\OneDrive\Desktop\datasets\tmdb_5000_credits.csv') 
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(2)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert4(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 4:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'].apply(convert4)
```

```python
movies['cast'] = movies['cast'].apply(convert4)
movies.head()
```

```python
movies.iloc[0].cast
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast'])
#new.head()
```

```python
new['crew'] = new['crew'].apply(lambda x: " ".join(x))
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(stem)
```

```python
new['tags'][3]
```

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
    
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
new[new['title'] == 'Avatar'].index[0]
```

```python
new.iloc[214].title
```

```python
def recommend(movie):
    movie_index = new[new['title'] == movie].index[0]
    distances= similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(new.iloc[i[0]].title)
        
```

```python
def recommend_director(director):
    movie_index = new[new['crew'] == director].index[0]
    distances= similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```

```python
recommend_director('ChristopherNolan')
```
